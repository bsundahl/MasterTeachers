```python
# Open files with positive and negative words
positive_file = open('positive-words.txt', 'r')
negative_file = open('negative-words.txt', 'r')

# Create list of positive and negative words
positive = []
negative = []

# Populate positive word list
for line in positive_file:
    line = line.strip()
    if len(line) > 0 and not line.startswith(";"):
        positive.append(line)

# Populate negative word list
for line in negative_file:
    line = line.strip()
    if len(line) > 0 and not line.startswith(";"):
        negative.append(line)
        
# Always close open files
positive_file.close()
negative_file.close()
```

```python
# My hashtag to search
myHashtag = '#brooklynnets -filter:retweets -filter:replies'

# List of tweets with a geotag
#geo_tweets = []

# List with positive word counts of each tweet
pos_word_count = []

# List with negative word count of each tweet
neg_word_count = []

# Search the twitter API
matchedTweets = twitter.search(q=myHashtag, count=100, lang='en')#, geocode="40.9257,-73.1409,10mi")
```

```python
# Lets look at the text of the tweets
i = 0
for tweet in matchedTweets["statuses"]: 
    
        # New tweet, so increase size of neg_word_count and
        # pos_word_count
        neg_word_count.append(0)
        pos_word_count.append(0)
        
        # Save tweet text in proper format
        text = tweet['text'].encode('utf-8')
    
        print "Tweet Num: ", i
        print "User    : ", tweet['user']['screen_name']
        print "Location: ", tweet['geo']
        print "Text    : ", text
        print ""
        #if tweet['geo'] != None:
        #    geo_tweets.append(tweet)
        

        
        # Check if words are positive
        for word in positive:
            
            searchPattern = re.compile("\\b" + word + "\\b")
            searchMatches = searchPattern.findall(text)
            pos_word_count[-1] += len(searchMatches)
            
        # Check if words are negative
        for word in negative:
            
            searchPattern = re.compile("\\b" + word + "\\b")
            searchMatches = searchPattern.findall(text)
            neg_word_count[-1] -= len(searchMatches)
            
        i += 1
```
