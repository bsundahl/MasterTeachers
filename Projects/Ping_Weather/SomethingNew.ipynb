```python
%matplotlib inline

from mpl_toolkits.basemap import Basemap, cm
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset as NetCDFFile
```

```python
SB_lat = 40.92
SB_long = 287
change = 20

m = Basemap(llcrnrlat=SB_lat-change,urcrnrlat=SB_lat+change,
            llcrnrlon=SB_long-change,urcrnrlon=SB_long+change)
m.drawcoastlines()
m.drawcountries()
```

```python
m.drawcountries()
m.drawcoastlines()
m.fillcontinents(color='green',lake_color='blue')
m.drawmapboundary(fill_color='blue')
m.drawparallels(np.arange(-80,80,5));
m.drawmeridians(np.arange(-180,180,5));
```

```python
m = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
m.bluemarble()
plt.show()
```
