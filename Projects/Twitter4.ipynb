```python
# My hashtag to search
myHashtag = '# -filter:retweets -filter:replies'

# Things we care about
avg_word_length = []
num_hashtags = []

# Search the twitter API
matchedTweets = twitter.search(q=myHashtag, count=100, lang='en')#, geocode="40.9257,-73.1409,10mi")
```

```python
# Lets look at the text of the tweets
i = 0
for tweet in matchedTweets["statuses"]: 
           
        # Save tweet text in proper format
        text = tweet['text'].encode('utf-8')
    
        print "Tweet Num: ", i
        print "User    : ", tweet['user']['screen_name']
        print "Location: ", tweet['geo']
        print "Text    : ", text
        print ""
        
        # Parse the text now
        text = text.split()
        
        # Calculate average word length
        num_words = 0
        num_chars = 0
        num_tags = 0
        
        for word in text:
            
            # Don't count hashtags
            if not word.startswith("#") and not word.startswith("http"):
                num_words += 1
                num_chars += len(word)
            elif word.startswith("#"):
                num_tags += 1
                
        # Add average to list
        avg_word_length.append(float(num_chars)/num_words)
        num_hashtags.append(num_tags)
            
        i += 1
```

```python
# Make a plot
fig, ax = plt.subplots()
ax.scatter(avg_word_length, num_hashtags)

# add some text for labels, title and axes ticks
ax.set_ylabel('Number of hastags')
ax.set_xlabel('Average Word Length')
ax.set_title('Something')
plt.axhline(y=0, color='k')

# 100 points from 0 word length to some fairly long word length
horizontal = numpy.linspace(0, 20, 100)
# Fitting a "y = mx + b" equation
m, b = numpy.polyfit(avg_word_length, num_hashtags, 1)
vertical = m * horizontal + b
ax.plot(horizontal, vertical)

plt.show()
```
