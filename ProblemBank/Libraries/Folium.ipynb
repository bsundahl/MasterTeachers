```python
from folium import plugins
import numpy as np

data = (np.random.normal(size=(100, 3)) *
        np.array([[1, 1, 1]]) +
        np.array([[48, 5, 1]])).tolist()

mapa = folium.Map([48., 5.], tiles='stamentoner', zoom_start=6)
mapa.add_children(plugins.HeatMap(data))
mapa
```

```python
mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12, tiles='Stamen Terrain')

icon = folium.Icon(color='blue', icon_color='white', icon='cloud',
                   angle=0, prefix='glyphicon')
p0 = folium.Marker([45.3788, -121.6625], popup='Mt. Hood Meadows', icon=icon)

icon = folium.Icon(color='green')
p1 = folium.Marker([45.3711, -121.7113], popup='Timberline Lodge', icon=icon)

p2 = folium.Marker([45.3700, -121.6823], popup='simple')

mapa.add_children(p0)
mapa.add_children(p1)
mapa.add_children(p2)
mapa
```
