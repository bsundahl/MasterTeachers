```python
def anagram(mystring):
    n = len(mystring)

    #initialize starting variables
    oldlist = [] #Will hold all of the anagrams
    newlist = [] #a temporary list that will expand upon oldlist inside the loop
    for i in range(n):
        oldlist.append(mystring[i]) #every string in oldlist now has one letter
    
    #combinatorially increase the length of the strings (and also the length of the list)
    for i in range(n-1): #n-1 because we already put one letter in each string (when we made oldlist)
        for partial in oldlist: #each string in oldlist is a partial anagram
            
            #make a list containing all of the letters in mystring that aren't in partial, accounting for duplicates
            workingstring = ""
            temp = partial
            for j in mystring: #j is the letter in mystring that we're considering
                if j in temp:
                    for k in range(len(temp)): #if j is already in our partial, remove one j from our partial.
                                                #this accounts for duplicate letters.
                        if j == temp[k]:
                            temp = temp[:k] + temp[k+1:]
                            break
                else:
                    workingstring = workingstring + j
            workinglist = list(workingstring) #our list is finished, now let's make new anagram partials
            
            
            #now that that's done, create a bunch of new strings that begin with the same "partial" anagram from
            #oldlist, but each have a different letter appended to the end.
            for j in workinglist:
                newlist.append(partial+j)
        
        #keep our loop going by reassigning names
        oldlist = newlist
        newlist = []
        
    return list(set(oldlist))
        
    
```
