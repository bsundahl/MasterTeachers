If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
threes = [x for x in range(10000) if x % 3 == 0]
print threes
```

```python
fives = [ x for x in range(10000) if x % 5 == 0]
print fives
```

```python
multiples = list(set(threes + fives))
```

```python
sum(multiples)
```

```python
def euler1(n, a, b):
    mysum = 0.0
    
    i = 0
    while i < n:
        if i % a == 0 and i % b != 0:
            mysum += i
        elif i % b == 0 and i % a != 0:
            mysum += i
        i += 1
            
    return mysum
```

```python
euler1(10, 3, 5)
```
